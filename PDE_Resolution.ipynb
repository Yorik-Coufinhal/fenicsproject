## Rapport de Project

### Introduction au problème

Nous avons décidé de modéliser l'action d'un poids sur une membrane (par exemple la toile d'un tambour, la simulation d'un matériau flexible pour l'industrie etc.) à l'aide de la librairie [FEniCS](https://fenicsproject.org/) [[1]](#[1]). Nous nous sommes cantonnés aux domaines de problèmes elliptiques et nous avons donc représenté le système dans un état stationnaire (i.e. statique, une fois l'équilibre atteint). Ce système est modélisé par l'[équation de Poisson](https://fr.wikipedia.org/wiki/%C3%89quation_de_Poisson) $-T\Delta D=p$ ($p$ la source, $D$ l'affaissement de la membrane, $T$ la tension de la membrane) [[2]](#[2])[[3]](#[3]). Le domaine pouvant assumer des formes diverses et variées, nous avons implémenté plusieurs topologies de domaine :

1. Un disque plein

2. Un disque avec trous alignés sur une maille

3. Un rectangle

4. Une ellipse

5. Un disque avec un trou circulaire, un trou rectangulaire

De plus, nous avons aussi implémenté deux familles de conditions aux bords : [[4]](#[4])

1. Les [conditions aux bords de Dirichlet](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Dirichlet)  (réflexion totale de l'onde par le bord)

2. Les [conditions aux bords de Robin](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Robin) (absorption partielle de l'onde par le bord)

<b>Notons que dans notre convention d'écriture, un affaissement $D$ négatif signifie la présence d'une dépression dans la membrane ; un affaissement $D$ positif signifie la présence d'une bosse dans la membrane ($D$ de l'anglais « [displacement](https://www.sciencedirect.com/topics/engineering/membrane-displacement) »).</b> Car en pratique les paramètres expérimentaux ne sont pas entièrement contrôlés, nous avons étudié l'impact des fluctuations expérimentales en générant des ensembles de paramètres autour de leur valeur théorique souhaitée à partir de [distributions Gaussiennes](https://fr.wikipedia.org/wiki/Loi_normale) d'écart-type propre à chaque paramètre.

### Algorithmes et équations

Revenons sur l'écriture de l'[équation de Poisson](https://fr.wikipedia.org/wiki/%C3%89quation_de_Poisson) :<br><br>

<center>$-T\Delta D=p\Longleftrightarrow\Delta D=-\frac{p}{T}\quad$ sur  $\Omega$</center>

avec $\Omega$ le domaine (i.e. la membrane), $T$ la tension de la membrane, $D$ l'affaissement et p la source d'application du poids sur la membrane. On pourrait [adimensioner](https://tel.archives-ouvertes.fr/tel-00995603v2/document) l'équation pour généraliser la résolution numérique (et utiliser les facteurs d'adimensionalisation pour retrouver une grandeur physique). On décide de ne pas le faire ici pour des questions de clarté du code.

On peut, entre autres, définir la source $p$ de poids appliqué à la membrane comme une [Gaussienne 2D](https://fr.wikipedia.org/wiki/Loi_normale_multidimensionnelle) d'expression :<br><br>

<center>$p=A\exp\{-((\frac{x-x_{0}}{\sigma})^{2}+(\frac{y-y_{0}}{\sigma})^{2})\}\quad$ sur $\Omega$</center>

avec $A$ un coefficient d'amplitude de la source, $x_{0}$ et $y_{0}$ les translations du point central d'application du poids par rapport à l'origine des axes et $\sigma$ l'écart-type de la [distributions Gaussiennes](https://fr.wikipedia.org/wiki/Loi_normale). On se sert de $A$ pour absorber les différents facteurs venant de la normalisation usuelle de la [distributions Gaussiennes](https://fr.wikipedia.org/wiki/Loi_normale) et ainsi simplifier l'écriture. 

Les [conditions aux bords de Dirichlet](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Dirichlet)  sont les suivantes : [[5]](#[5])[[6]](#[6])<br><br>

<center>$D=\alpha\quad$ sur $\partial\Omega$</center>

avec $D$ l'affaissement de la membrane, $\alpha$ la valeur de l'affaissement sur le contour du domaine (i.e. de la membrane) et $\partial\Omega$ le contour du domaine. Ces [https://fr.wikipedia.org/wiki/Condition_aux_limites||conditions aux bords] sont implémentées par la fonction [DirichletBC](https://fenicsproject.org/olddocs/dolfin/2017.2.0/python/programmers-reference/fem/bcs/DirichletBC.html) de [FEniCS](https://fenicsproject.org/). 

Les [conditions aux bords de Robin](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Robin) sont les suivantes : [[4]](#[4])[[5]](#[5])<br><br>

<center>$a\frac{\partial D}{\partial n}+bD=\alpha\quad$ sur $\Omega$</center>

avec $a$ et $b$ deux coefficients d'amplitude de chaque terme et $n$ le [champ de normales](https://fr.wikipedia.org/wiki/Normale_%C3%A0_une_surface#:~:text=Un%20champ%20de%20normales%20(normales,est%20appel%C3%A9e%20application%20de%20Gauss.) lié au bord. Les [conditions aux bords de Robin](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Robin) ne sont pas implémentées par [FEniCS](https://fenicsproject.org/), nous les avons implémentées en passant par la [formulation variationnelle](https://fr.wikipedia.org/wiki/Formulation_faible) de l'équation différentielle : [[5]](#[5])[[6]](#[6])[[7]](#[7])<br><br>

<center>$\int_{\Omega}\nabla D\cdot\nabla vdx+\alpha\int_{\partial\Omega}D\cdot vds=-\int_{\Omega}\frac{p}{T}vdx$</center>

avec $v$ une fonction de test et $\alpha$ l'amplitude des [conditions aux bords de Robin](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Robin) (différent du \alpha des [conditions aux bords de Dirichlet](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Dirichlet) ). Le signe de $\alpha$ détermine si les bords sont puits ou sources de poids dans la [solution stationnaire](https://fr.wikipedia.org/wiki/%C3%89tat_stationnaire) de notre simulation. [[4]](#[4])

Ainsi, la simulation numérique procède de la façon suivante : [[6]](#[6])[[7]](#[7])

1. On initialise les variables de la simulation (amplitudes, topologie du domaine, liste des variables à échantillonner etc.)

2. On génère la grille de points [FEniCS](https://fenicsproject.org/) (mesh) à partir de la topologie de notre domaine.

3. On commence la boucle sur le nombre d'itération de notre simulation :

    (a) On extrait les variables de la simulation.

    (b) On résout l'[équation différentielle](https://fr.wikipedia.org/wiki/%C3%89quation_de_Poisson) avec [conditions aux bords](https://fr.wikipedia.org/wiki/Condition_aux_limites) en utilisant les fonctions de [FEniCS](https://fenicsproject.org/).

    (c) On exporte l'amplitude d'affaissement maximale et la liste des paramètres échantillonnés dans une liste results pour chaque itération de notre simulation.

Ce qui vient en fin de section 4 correspond à des morceaux de code utiles pour la visualisation, notamment, le graphe de la source $\frac{p}{T}$ (gauche) et les $n$ graphes de l'affaissement (droite) sur le domaine à chaque $i$ème itération de la simulation ($i$ la période de calcul des graphes, $n$ le nombre de [conditions aux bords](https://fr.wikipedia.org/wiki/Condition_aux_limites) analysées par simulation). En section 5, on effectue une analyse statistique des différentes variables échantillonnées pour chaque [conditions aux bords](https://fr.wikipedia.org/wiki/Condition_aux_limites).

### Description des Résultats
Présentons les résultats de nos simulations (sauf mentionné, on se place dans des conditions aux bords de Dirichlet avec $\alpha = 0$) :

#### Echantillonnage de l'excentricité $r=\sqrt{x_0^2 + y_0^2}$ de la source de poids $p$ par rapport au domaine $\Omega$

L'image qui suit représente l'évolution du maximum d'affaissement $D_{max}$ en fonction de la distance radiale $r=\sqrt{x_{0}^{2}+y_{0}^{2}}$ du centre d'application de la source dans le domaine. On observe que $|D_{max}|$ diminue jusqu'à $D_{max}=0$ selon une [Gaussienne](https://fr.wikipedia.org/wiki/Loi_normale) $\mu e^{-\theta r^{2}}$ avec $\mu$ et $\theta$ des paramètres spécifiques à la simulation. Ceci est logique quand on considère que géométrie de la source est une [Gaussienne 2D](https://fr.wikipedia.org/wiki/Loi_normale_multidimensionnelle).

![title](img/r.png)

#### Echantillonnage de la tension $T$ de la membrane

Pour le cas $T>0$, la relation entre $T$ et $D_{max}$ suit une [distribution de Pareto](https://fr.wikipedia.org/wiki/Loi_de_Pareto) de la forme $D_{max}(T)=\mu(1-e^{\theta T})$ avec $\mu$ et $\theta$ des paramètres spécifiques à la simulation. La raison de ce comportement provient du fait que $\Delta D_{max}\propto\frac{1}{T}$. La tension de la membrane et l'affaissement sont donc inversement proportionnels. Pour le cas $T \leq 0$, la tension change le signe de l'affaissement (bosse plutôt que creux). On ne considère pas ce dernier cas ($D_{max}>0$) dans cette analyse.

![title](img/T.png)

#### Echantillonnage de l'amplitude $A$ de la source de poids $p$

L'affaissement $D_{max}$ est linéairement proportionnel à l'amplitude du poids de part le fait que $\Delta D \propto p=Ae^{\nu}$. Plus l'amplitude $A$ est grande, plus $|D_{max}|$ est important.

![title](img/A.png)

#### Echantillonnage de l'écart-type $\sigma$ de la distribution Gaussienne comme source de poids $p$

La relation entre $D_{max}$ et $\sigma$ l'« étalement » de la source de poids $p$ suit une [distribution Lorentzienne](https://fr.wikipedia.org/wiki/Loi_de_Cauchy_(probabilit%C3%A9s)) du type $D_{max}(\mu)=\mu\left[\frac{\gamma^{2}}{(\sigma-\sigma_{0})^{2}+\gamma^{2}}\right]$ avec $\mu_{0}$, $\gamma$ et $\mu$ des paramètres spécifiques à la simulation. On interprète la relation entre ces deux par le fait que plus $\sigma \rightarrow 0$, plus la [masse](https://fr.wikipedia.org/wiki/Mesure_finie#Fonctions_int%C3%A9grables) de la source de poids sur le domaine sera petite, donc aussi l'amplitude d'affaissement $|D_{max}|$.

![title](img/sigma.png)

#### Echantillonnage du coefficient $\alpha$ pour les conditions aux bords de Dirichlet

La topologie de l'affaissement $D_{max}$ est indépendante des [conditions aux bords de Dirichlet](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Dirichlet) . $\alpha$ n'est n'est que le facteur qui règle la hauteur de la membrane donc la relation entre $D_{max}$ et $\alpha$ est $D_{max}(\alpha)\propto\alpha$.

![title](img/alpha_D.png)

#### Echantillonnage du coefficient $\alpha$ pour les conditions aux bords de Robin

On observe que la relation entre $D_{max}$ et $\alpha$ pour les [conditions aux bords de Robin](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Robin) est de la forme $D_{max}(\alpha)\propto\frac{1}{\alpha}+\beta$ avec $\beta$ une constante de translation de l'[hyperbole](https://fr.wikipedia.org/wiki/Hyperbole_(math%C3%A9matiques)). $\alpha=0$ est une [singularité](https://fr.wikipedia.org/wiki/Singularit%C3%A9). Pour $\alpha > 0$ (resp. $\alpha < 0$), plus $\alpha\rightarrow\infty$ (resp. $\alpha\rightarrow-\infty$), plus l'affaissement $D_{max}$ sature vers la constante $\beta$ (l'effet de transmission de la source en dehors du domaine par les bords sature [[4]](#[4])).

![title](img/alpha_R.png)


#### Exemple de l'impact de la topologie du domaine et des conditions aux bords sur l'affaissement

La deuxième ligne de graphe (topologie de disques pleins contenant une grille de trous) est la plus parlante : on voit nettement l'action des [conditions aux bords](https://fr.wikipedia.org/wiki/Condition_aux_limites) sur l'affaissement de la membrane pour une source [Gaussienne 2D](https://fr.wikipedia.org/wiki/Loi_normale_multidimensionnelle) : [[4]](#[4])

- Les [conditions aux bords de Dirichlet](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Dirichlet) marquent une diminution de l'amplitude d'affaissement $|D|$ très nette autour des bords du domaine mais la [masse](https://fr.wikipedia.org/wiki/Mesure_finie#Fonctions_int%C3%A9grables) de la source a été préservée.

- Les [conditions aux bords de Robin](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Robin) marquent la transmission partielle de la source $\frac{p}{T}$ vers l'extérieur du domaine de la membrane trouées (une impédance à l'affaissement $D$). On remarque en effet la géométrie en « trèfle » de l'affaissement $D$ au centre de la grille de trous, bien que comparé aux [conditions aux bords de Dirichlet](https://fr.wikipedia.org/wiki/Condition_aux_limites_de_Dirichlet), les bords du domaine ont un affaissement $D$ non fixé. Ainsi,  la [masse](https://fr.wikipedia.org/wiki/Mesure_finie#Fonctions_int%C3%A9grables) de la source n'a ici pas été préservée.

![title](img/topologies.png)

### Conclusion

Modéliser l'action d'une source sur un domaine est courant [en Physique et en Ingéniérie](https://project.inria.fr/mastic/files/2013/09/talk-JL.pdf) (garantir la fiabilité du matériau aux conditions dont il est exposé, ici, un poids sur une membrane ; modéliser l'action d'un faisceau chargé sur une enveloppe sensible au magnétisme etc.). Si la théorie peut permettre de considérer des cas parfaits (e.g. l'application d'une force en un point singulier), la pratique se veut plus incertaine. Prendre en compte cette incertitude dans les paramètres de notre simulation permet ainsi d'étudier la réponse du matériau autour d'une plage d'incertitude simulant les conditions expérimentales. En étudiant la relation entre un paramètre (point d'application de la force $r$, tension de la membrane $T$ etc.) et l'affaissement $D$, nous pouvons extraire des relations entre grandeurs. Cela permet finalement de conseiller le client de manière plus judicieuse. Par exemple :
* si l'affaissement est à éviter
* qu'il existe une relation quadratique entre l'affaissement et une grandeur $A$
* une relation linéaire entre l'affaissement et une grandeur $B$

alors il sera potentiellement plus intéressant de dépenser une somme $X$ pour contrôler le paramètre expérimental $A$ plutôt que le paramètre $B$. Cependant, si la somme $X$ est très grande comparée à la somme $Y$ à dépenser pour modifier $B$, alors ne vaudrait-il pas mieux jouer sur le paramètre $B$ ?<br>

Il est ici possible d'étudier beaucoup plus de cas :
- Source non Gaussienne
- Tension inhomogène de la membrane
- Conditions aux Bords différentes selon l'endroit dans le domaine
- etc.

## Initialisation

```python
from fenics import *
from mshr import *
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
from matplotlib import cm
from tqdm.notebook import tqdm

#####
# Initialize variables
#####

iteration   = 100           # Nombre d'échantillons pour les statistiques
plot_period = 45            # Nombre de simulations entre chaque plots
cmap_name   = "magma"       # La colormap utilisée pour les plots à venir
shape       = "Circle"      # "Grid" "Circle" "Rectangle" "Ellipse" "Holed"
RadiusDiv   = 2             # Diviseur du rayon des trous dans la mesh (si trous)
cdt_bords   = ["Dirichlet", # Liste des Conditions aux Bords à étudier: ["Dirichlet", "Robin"]
               "Robin"]

Base = {# Nom       : (Variable, ecart-type)
        "$x_0$"     : (0., 1),     # Shift du point d'application du load sur l'axe des x
        "$y_0$"     : (0., 1),     # Shift du point d'application du load sur l'axe des y
        "$A$"       : (10, 3),     # Facteur devant l'exponentielle du load
        "$\\sigma$" : (0.5, 3),    # Standard deviation dans l'exponentielle du load
        "$T$"       : (1000, 800), # Tension
        "$\\alpha$" : (10, 5)}     # Coefficient des conditions aux bords (Dirichlet + Robin)

ToSample = [# "Nom"
            "$x_0$",
            "$y_0$",
#             "$A$",
#             "$\\sigma$",
#             "$T$",
#             "$\\alpha$"
            ]

#####
# Sampling setup
#####

NewParams = {} # Dictionnaire contenant les paramètres échantillonnés aléatoirement

for key, val in Base.items():
    if key in ToSample: # variable centrée sur val[0], d'écart-type val[1]
        NewParams[key] = [rd.normal(val[0], val[1]) for i in range(iteration)]
    else:
        NewParams[key] = [val[0] for i in range(iteration)]

# Note: Attentions aux cas anormaux de valeurs négatives pour des grandeurs
# purement positives (Sigma etc.). Le choix a été fait de ne pas prendre la
# valeur absolue pour éviter la perte de la forme de la distribution dans la
# génération des échantillons aléatoires. Il revient à l'utilisateur de prendre
# ces valeurs anormales en compte en les supprimant de l'analyse post-simulation.
```

## Mesh Processing

```python
#####
# Mesh Processing
#####

if shape == "Grid" or shape == "Circle":
    domain = Circle(Point(0, 0), 1)
elif shape == "Rectangle":
    domain = Rectangle(Point(-1, -1),Point(1, 1))
elif shape == "Ellipse":
    domain = Ellipse(Point(0, 0), 1.25, 0.75)
elif shape == "Holed":
    domain = Circle(Point(0, 0), 1) \
                - Rectangle(Point(-0.05, -0.05), Point(-0.15, -0.15)) \
                - Circle(Point(0.3, 0.35), 0.1)
if shape == "Grid":
    x = np.linspace(-1,1,10)
    xv, yv = np.meshgrid(x, x)
    length = len(x)
    radius = (x[1] - x[0])/2/RadiusDiv
    
    z = xv*xv + yv*yv <= 1-radius*2

    for i in tqdm(range(length)):
        for j in range(length):
            if z[i][j]:
                domain -= Circle(Point(x[i], x[j]), radius)

mesh = generate_mesh(domain, 64)
```

## Résolution de l'EDP

```python
results = [[] for i in range(len(cdt_bords))] # Préparation de la liste de résultats

# Préparation des colormaps à utiliser pour les plots
cmap   = cm.get_cmap(cmap_name)        # Colormap normale (source)
cmap_r = cm.get_cmap(cmap_name + "_r") # Colormap inversée (affaissement)

# On refait la simulation "iteration" fois
for i in tqdm(range(iteration)):
    for j, BC in enumerate(cdt_bords):
        #####
        # Variable extraction
        #####
        sigma = NewParams["$\\sigma$"][i]
        x0    = NewParams["$x_0$"]    [i]
        y0    = NewParams["$y_0$"]    [i]
        A     = NewParams["$A$"]      [i]
        T     = NewParams["$T$"]      [i]
        alpha = NewParams["$\\alpha$"][i]

        #####
        # FENICS PDE definition
        #####
        p = Expression('(A/T)*exp(-(pow((x[0] - x0)/sigma, 2) + pow((x[1] - y0)/sigma, 2)))',
                       degree=1, sigma=sigma, x0=x0, y0=y0, A=A, T=T)

        V = FunctionSpace(mesh, 'P', 1)

        w = TrialFunction(V)
        v = TestFunction(V)

        #####
        # PDE solving
        #####
        if BC == "Robin":
            F    = (inner(grad(w), grad(v))*dx + inner(alpha*w, v)*ds - inner(-p, v)*dx)
            a, L = lhs(F), rhs(F)
            w    = Function(V)

            solve(a == L, w)

        else:
            def boundary(x, on_boundary):
                return on_boundary

            u_D = Constant(alpha)
            bc  = DirichletBC(V, u_D, boundary)

            a   = dot( grad(w), grad(v) ) * dx
            L   = -p*v*dx
            w   = Function(V)

            solve(a == L, w, bc)

        #####
        # Source interpolation
        #####
        p=interpolate(p,V)

        #####
        # Result export
        #####
        results[j].append((min(np.array(w.vector())), [NewParams[key][i] for key in ToSample]))

    #####
    # Periodic rough plotting
    #####
        # Plot of D according to boundary condition
        if i % plot_period == 0:
            plt.figure(1, figsize=(20,5))
            
            plt.subplot(1,3,j+2)
            d=plot(w, mode='color', cmap=cmap_r,
                   title="$D$ avec cdts aux bords de " + cdt_bords[j])
            plt.colorbar(d)

    # Plot of source of load divided by tension
    if i % plot_period == 0:
        plt.subplot(1,3,1)
        c=plot(p, title='Amplitude de source $\\frac{p}{T}$ de poids', mode='color', cmap=cmap)
        plt.colorbar(c)

        plt.show()
```

## Analyse Statistique

```python
#####
# Statistical Analysis
#####

for k, BC in enumerate(cdt_bords):
    dists = np.zeros(iteration)

    for i in range(len(results[k][0][1])):
        variables   = []
        deflections = []

        # i = index de la quantité (e.g. 1: x0, 2: y0...)
        # j = index de l'itération (e.g. 1: premiere simulation, 2: ...)
        # res[0] = déflection maximale (scalaire)
        # res[1] = array des paramètres choisis aléatoirement
        for j, res in enumerate(results[k]):
            deflections.append(res[0])
            variables.append(res[1][i])

            # Si on est dans le cas de l'étude de l'effet
            # de la translation de la source dans le domaine
            # on s'intéresse à regarder la distance euclidienne
            # d = sqrt((x_0-x_centre)^2 + (y_0-y_centre)^2) = sqrt(x_0^2 + y_0^2)
            if ToSample == ["$x_0$", "$y_0$"]:
                dists[j] += res[1][i] * res[1][i]

        plt.title("Affaissement $D_{max}$ comme fonction de " + ToSample[i] + \
                  "\nen cdt aux bords de " + cdt_bords[k])
        plt.scatter(variables, deflections)
        plt.xlabel(ToSample[i])
        plt.ylabel("$D$", rotation=0)
        plt.grid()

        plt.show()

    # Visualisation spéciale dans le cas où on étudie l'effet
    # de la translation de la source dans le domaine
    if ToSample == ["$x_0$", "$y_0$"]:
        dists = np.sqrt(dists)
        plt.title("Affaissement $D_{max}$ de la membrane en fonction de\nla distance $r$ de la source" +
                  "par rapport au centre du domaine\nen cdts aux bords de " + cdt_bords[k])
        plt.scatter(dists, np.array(deflections)*T/A)
        plt.grid()
        plt.xlabel("$r$")
        plt.ylabel("$D$", rotation=0)
        plt.xlim(0, max(dists))

        plt.show()
```
